```python
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.mixture import GaussianMixture
from sklearn.datasets.samples_generator import make_blobs
```

```python
# Create some dataset
X,y = make_blobs(cluster_std=2, random_state=20, n_samples=1000, centers=2)
# X is a matrix with 2 columns - the 2D coordinates
# y is a cluster label; you don't care about this as you don't have any labels
```

```python
# Plot the made up clusters
plt.scatter(X[:,0], X[:,1], s=5, c=[.9, .9, .9]);
```

```python
# Instantiate and fit the model
GMM = GaussianMixture(n_components=2).fit(X)
# Check if the fit has converged
print('Converged:',GMM.converged_)           
```

```python
# Plot the cluster centers to see if they make sense
plt.scatter(X[:,0], X[:,1], s=5, c=[.9, .9, .9])
plt.scatter(GMM.means_[:,0], GMM.means_[:,1], s=10, c='red');
```

```python
# Classify your points into 2 clusters
pred_prob = GMM.predict_proba(X)
pred_class = pred_prob.argmax(axis = 1)
```

```python
# Plot the two predicted classes
plt.scatter(X[:,0], X[:,1], s=5, c=pred_class);
```

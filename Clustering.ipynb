# Clustering

```python
import numpy as np
from sklearn import datasets

# Load a dataset
X = datasets.load_iris()['data']
```

```python
# Plot a scatter of 2 features
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(X[:,0], X[:,1], 'bo')
plt.xlabel('Feature 0')
plt.ylabel('Feature 1');
```

```python
# Make 2 clusters using k-means
from sklearn.cluster import KMeans

# Fit cluster
model = KMeans(n_clusters=2).fit(X[:,:2])

# Predict cluster type
label = model.predict(X[:,:2])

# Plot clusters
plt.plot(X[label==0, 0], X[label==0, 1], 'ro')
plt.plot(X[label==1, 0], X[label==1, 1], 'bo')
plt.xlabel('Feature 0')
plt.ylabel('Feature 1');
```

```python
# Make 3 clusters using k-means
from sklearn.cluster import KMeans

# Fit cluster
model = KMeans(n_clusters=3).fit(X[:,:2])

# Predict cluster type
label = model.predict(X[:,:2])

# Plot clusters
plt.plot(X[label==0, 0], X[label==0, 1], 'ro')
plt.plot(X[label==1, 0], X[label==1, 1], 'bo')
plt.plot(X[label==2, 0], X[label==2, 1], 'go')
plt.xlabel('Feature 0')
plt.ylabel('Feature 1');
```

### Hierarchical clustering

```python
from sklearn.cluster import AgglomerativeClustering

label = AgglomerativeClustering(n_clusters=3, 
                                affinity='euclidean', 
                                linkage='ward').fit_predict(X[:,:2])

# Plot clusters
plt.plot(X[label==0, 0], X[label==0, 1], 'ro')
plt.plot(X[label==1, 0], X[label==1, 1], 'bo')
plt.plot(X[label==2, 0], X[label==2, 1], 'go')
plt.xlabel('Feature 0')
plt.ylabel('Feature 1');
```

For ploting the corresponding dendrogram, the scipy implementation of hierarchical/agglomerative clustering is useful.

```python
from scipy.cluster.hierarchy import dendrogram, linkage

# Perform hierarchical clustering
Z = linkage(X[:15,:2])

# Plot the hierarchical clustering as a dendrogram
dendrogram(Z);
```
